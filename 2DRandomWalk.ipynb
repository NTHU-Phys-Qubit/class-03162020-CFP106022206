```python
import numpy as np
import matplotlib.pyplot as plt

step = 1000
pos = np.zeros([step,2]) #

for i in range(1,step):
    theta = np.random.random() * 2 * np.pi
    pos[i,0] = pos[i-1,0] + np.cos(theta)
    pos[i,1] = pos[i-1,1] + np.sin(theta)
plt.style.use('dark_background')
plt.plot(pos[:,0], pos[:,1])
```
